```python
from collections import deque

history = []
download_queue = deque()

total_kunjungan = 0
download_selesai = 0


def kunjungi_halaman():
    global total_kunjungan

    url = input("Masukkan URL halaman: ")
    history.append(url)
    total_kunjungan += 1

    print(f"Berhasil mengunjungi: {url}")


def back_halaman():
    if len(history) == 0:
        print("History kosong.")
    else:
        halaman = history.pop()
        print(f"Kembali dari halaman: {halaman}")


def tampilkan_history():
    if len(history) == 0:
        print("Tidak ada history.")
    else:
        print("\nHISTORY")

        for i in range(len(history)-1, -1, -1):
            print(history[i])


def tambah_download():
    file = input("Masukkan nama file: ")
    download_queue.append(file)

    print(f"File '{file}' ditambahkan ke antrian download.")


def proses_download():
    global download_selesai

    if len(download_queue) == 0:
        print("Tidak ada antrian download.")
    else:
        file = download_queue.popleft()
        download_selesai += 1

        print(f"Download selesai: {file}")


def tampilkan_download():
    if len(download_queue) == 0:
        print("Antrian download kosong.")
    else:
        print("\nANTRIAN DOWNLOAD")
        for file in download_queue:
            print(file)



def statistik():
    print("\nSTATISTIK SISTEM")
    print(f"Jumlah halaman pernah dikunjungi : {total_kunjungan}")
    print(f"Jumlah halaman di history        : {len(history)}")
    print(f"Jumlah file selesai download    : {download_selesai}")
    print(f"Jumlah file menunggu download   : {len(download_queue)}")


while True:
    print("\nMENU")
    print("1. Kunjungi Halaman")
    print("2. Back")
    print("3. Tampilkan History")
    print("4. Tambah Download")
    print("5. Proses Download")
    print("6. Tampilkan Queue Download")
    print("7. Statistik")
    print("0. Keluar")

    pilihan = input("Pilih menu: ")

    if pilihan == "1":
        kunjungi_halaman()

    elif pilihan == "2":
        back_halaman()

    elif pilihan == "3":
        tampilkan_history()

    elif pilihan == "4":
        tambah_download()

    elif pilihan == "5":
        proses_download()

    elif pilihan == "6":
        tampilkan_download()

    elif pilihan == "7":
        statistik()

    elif pilihan == "0":
        print("Program selesai. Terima kasih!")
        break

    else:
        print("Pilihan tidak valid.")
```
