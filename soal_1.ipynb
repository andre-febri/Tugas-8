```python
stack_buku = []

n = int(input("masukkan jumlah buku: "))

for i in range(n):
  judul = input(f"judul buku ke-{i+1}: ")
  stack_buku.append(judul)

print("\nisi Stack:")
for buku in reversed(stack_buku):
  print(buku)

if stack_buku:
  buku_diambil = stack_buku.pop()
  print("\nbuku yang diambil: ", buku_diambil)

print("\nisi stack setelah pop:")
for buku in reversed(stack_buku):
  print(buku)
```
