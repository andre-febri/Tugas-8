```python
from collections import deque

antrian = deque()

n = int (input("masukkan jumlah pelanggan: "))

for i in range(n):
  nama = input(f"nama pelanggan ke-{i+1}: ")
  antrian.append(nama)

if antrian:
  dilayani = antrian.popleft()
  print("\npelanggan dilayani:", dilayani)

print("\nsisa antrean: ")
for pelanggan in antrian:
    print(pelanggan)
```
